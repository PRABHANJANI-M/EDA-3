```python
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'
if 'bmi' in df.columns:
    df['bmi_category'] = df['bmi'].apply(bmi_category)
df
```

Section 3 – Univariate Analysis

```python
numeric_vars = ['Age', 'Sleep Duration', 'Physical Activity Level']
df[numeric_vars].hist(figsize=(10,6))
plt.tight_layout()
plt.show()
```

```python
print(df['Sleep Duration'].mean())
print(df['Occupation'].mode())
```

```python
counts = df['Stress Level'].value_counts()
plt.bar(counts.index, counts.values, color='skyblue')
plt.xlabel("Stress Level")
plt.ylabel("Count")
plt.title("Distribution of Stress Levels")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#ff9999','#66b3ff']
)
plt.title('Gender Distribution')
plt.show()
```

```python
sleep_counts = df['Sleep Disorder'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(sleep_counts,labels=sleep_counts.index,autopct='%1.1f%%',startangle=90,colors=['#99ff99','#ffcc99','#66b3ff'])
plt.title('Sleep Disorder Distribution')
plt.show()
```

```python
df['Activity_Stress_Sum'] = df['Physical Activity Level'] + df['Stress Level']
df['Sleep_vs_Stress'] = df['Sleep Duration'] - df['Stress Level']
df['Steps_per_Activity'] = df['Daily Steps'] * df['Physical Activity Level']
df['Sleep_per_Age'] = df['Sleep Duration'] / df['Age']
df['High_HeartRate'] = (df['Heart Rate'] > 80).astype(int)
df
```

```python
subset = df[['Sleep Duration', 'Stress Level']]
correlation_matrix = subset.corr()
plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix,annot=True,cmap='cool',fmt=".2f",linewidths=0.5)
plt.title("Correlation Heatmap: Sleep vs Stress")
plt.show()
```
