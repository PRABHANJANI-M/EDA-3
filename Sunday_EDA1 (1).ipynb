```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
df
```

```python
from scipy import stats

# Select numerical columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

# ===== BEFORE OUTLIER REMOVAL =====
plt.figure(figsize=(12, 6))
df[num_cols].boxplot()
plt.title("Before Outlier Removal (Boxplot)")
plt.ylabel("Value Range")
plt.xticks(rotation=45)
plt.show()

# ===== OUTLIER REMOVAL USING Z-SCORE =====
z = np.abs(stats.zscore(df[num_cols]))
threshold = 3
df = df[(z < threshold).all(axis=1)] 

# ===== AFTER OUTLIER REMOVAL =====
plt.figure(figsize=(12, 6))
df[num_cols].boxplot()
plt.title("After Outlier Removal (Boxplot)")
plt.ylabel("Value Range")
plt.xticks(rotation=45)
plt.show()

# ===== CHANGE SUMMARY =====
print(f"Rows remaining after outlier removal: {len(df)}")
```

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'
if 'bmi' in df.columns:
    df['bmi_category'] = df['bmi'].apply(bmi_category)
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(30, 30))

for i, col in enumerate(df.columns, 1):
    plt.subplot(5, 3, i)
    if df[col].dtype in ['int64', 'float64']:
        plt.hist(df[col], bins=20, edgecolor='black')
        plt.xlabel(col)
        plt.ylabel("Count")
    else:
        df[col].value_counts().plot(kind='bar')
        plt.xlabel(col)
        plt.ylabel("Count")
    plt.title(f"Distribution of {col}")

plt.tight_layout()
plt.show()
# Select numeric columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate skewness
skew_values = df[num_cols].skew()

print("Skewness for all numeric columns:")
print(skew_values)

```

```python
df_norm = df.copy()
skewed_cols = ['Heart Rate', 'Daily Steps']

for col in skewed_cols:
    df_norm[col] = np.log1p(df_norm[col])
print("Skewness after normalization:")
print(df_norm[skewed_cols].skew())

plt.figure(figsize=(12,5))
for i, col in enumerate(skewed_cols, 1):
    plt.subplot(1, 2, i)
    sns.histplot(df_norm[col], kde=True, color='skyblue')
    plt.title(f'{col} After Log Transformation')
plt.tight_layout()
plt.show()
```

```python
# ----- Print Averages -----
avg_sleep = round(df['Sleep Duration'].mean(), 2)
avg_stress = round(df['Stress Level'].mean(), 2)
avg_quality = round(df['Quality of Sleep'].mean(), 2)

print("Average Sleep Duration:", avg_sleep, "hours")
print("Average Stress Level:", avg_stress)
print("Average Quality of Sleep:", avg_quality)

# ----- Plot Averages -----
plt.figure(figsize=(10,6))
averages = [avg_sleep, avg_stress, avg_quality]
labels = ['Sleep Duration', 'Stress Level', 'Quality of Sleep']

bars = plt.bar(labels, averages, color=['pink','skyblue','#74c476'], edgecolor='orange')

# Add values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, round(yval,2), ha='center', fontsize=11)

plt.title("Average Sleep Metrics", fontsize=14, fontweight='bold')
plt.ylabel("Average Value")
plt.show()
```

```python
male_sleep = df[df['Gender'] == 'Male']['Sleep Duration']
female_sleep = df[df['Gender'] == 'Female']['Sleep Duration']
t_stat, p_val = stats.ttest_ind(male_sleep, female_sleep)
print("T-statistic:", t_stat)
print("P-value:", p_val)
if p_val < 0.05:
    print("Significant difference in Sleep Duration between genders.")
else:
    print("No significant difference between genders.")
```

```python
sleep_gender_counts = df.groupby(['Gender', 'Sleep Disorder']).size().reset_index(name='Count')

# Plot grouped bar chart
plt.figure(figsize=(8,6))
sns.barplot(data=sleep_gender_counts, x='Gender', y='Count', hue='Sleep Disorder', palette='Set2')
plt.title('Sleep Disorder Distribution by Gender')
plt.ylabel('Number of People')
plt.show()
```
